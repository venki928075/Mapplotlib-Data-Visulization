```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
```

```python
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
 fig = plt.figure()
    
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('Set title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.15,0.4,0.3])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('first_post')

axes[1].plot(x,y)
axes[1].set_title('second_post')

plt.tight_layout() #if anytjing is overlapping
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(3,2))

axes[0].plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))

axes[0].plot(x,y, label='X Squared')

axes[1].plot(x,y, label='X cubed')
plt.tight_layout()
axes.legend()
```

```python
fig.savefig('my_pic.png', dpi=200)
```
